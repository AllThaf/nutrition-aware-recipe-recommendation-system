# Tahap 3: Build CBF Features
Tahap ini berfokus pada pengambilan data mentah dan pengekstraksian fitur teks (seperti nama masakan, ingredients, dan tags) menjadi satu dokumen utuh (`combined_text`).

```python
import pandas as pd
import pickle
from pathlib import Path
import sys

# Mendaftarkan direktori cbf agar modul lokal bisa dibaca
sys.path.append('.') 
from feature_extractor import load_recipes, extract_text_features

# Buat direktori penampungan output
OUT_DIR = Path("outputs/models")
OUT_DIR.mkdir(parents=True, exist_ok=True)
```

```python
print("1. Memuat Recipe Dataset Mentah...")
df_raw = load_recipes()
print(f"Ukuran dataset saat ini: {df_raw.shape}")
display(df_raw.head(3))
```

```python
print("2. Ekstraksi Fitur Teks ke dalam combined_text...")
df_processed = extract_text_features(df_raw)
print("Fitur teks selesai diekstrak.")
display(df_processed[['id', 'combined_text']].head(3))
```

```python
print("3. Menyimpan Hasil Ekstraksi Sementara...")
out_file = OUT_DIR / "cbf_features.pkl"
df_processed[['id', 'cf_idx', 'combined_text']].to_pickle(out_file)
print(f"File berukuran ringan tersimpan di: {out_file}")
```
